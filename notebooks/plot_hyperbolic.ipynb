```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# 定义 x 和 y 的范围
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
x, y = np.meshgrid(x, y)

# 计算 z 值，马鞍面公式：z = x^2 - y^2
z = x**2 - y**2
```

```python
# 创建三维图形对象
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 绘制曲面，使用自定义颜色映射
surf = ax.plot_surface(x, y, z, cmap='plasma', edgecolor='none', alpha=0.8)

# 移除坐标轴平面
# 移除背景平面
ax.xaxis.pane.fill = False  # 移除 x 轴平面
ax.yaxis.pane.fill = False  # 移除 y 轴平面
ax.zaxis.pane.fill = False  # 移除 z 轴平面

# 隐藏坐标轴线
# ax.w_xaxis.line.set_color((1.0, 1.0, 1.0, 0.0))  # 隐藏 x 轴线
# ax.w_yaxis.line.set_color((1.0, 1.0, 1.0, 0.0))  # 隐藏 y 轴线
# ax.w_zaxis.line.set_color((1.0, 1.0, 1.0, 0.0))  # 隐藏 z 轴线

# 隐藏刻度线和标签
# ax.set_xticks([])  # 去除 x 轴刻度
# ax.set_yticks([])  # 去除 y 轴刻度
# ax.set_zticks([])  # 去除 z 轴刻度

# 隐藏网格线
ax.grid(None)

ax.axis('off')
# 设置视角
ax.view_init(elev=30, azim=110)

plt.savefig('figs/hyperbolic.svg', dpi=600)
plt.savefig('figs/hyperbolic.eps', dpi=600)
plt.savefig('figs/hyperbolic.pdf', dpi=600)
plt.savefig('figs/hyperbolic.png', dpi=600)
# 显示图形
plt.show()
```
