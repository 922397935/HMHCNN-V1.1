#### 可自定义层内连接及控制连边概率

```python
small_hg1 = dhg.Hypergraph(5, [(0, 1, 2), (2, 3), (0, 4)])
small_hg2 = dhg.Hypergraph(5, [(1, 3, 2), (0, 3), (2, 4)])
small_mhg = MultilayerHypergraph(10, 2, layers_list=[small_hg1, small_hg2])
small_mhg = small_mhg.construct_multi_layer_hypergraph()
print(small_mhg)
```

```python
small_v_label = [str(i) for i in range(small_mhg.num_v)]
e_color = ['r'] * 3 + ['b'] * 3
v_color = ['r'] * 5 + ['b'] * 5
small_mhg.draw(v_label=small_v_label, v_color=v_color, e_color=e_color)
plt.savefig('figs/small_hg.svg', dpi=600)
plt.savefig('figs/small_hg.eps', dpi=600)
plt.savefig('figs/small_hg.pdf', dpi=600)
```

```python
small_mhg.add_hyperedges([[4, 7], [2, 5]])
print(small_mhg)
```

```python
small_v_label = [str(i) for i in range(small_mhg.num_v)]
e_color = ['r'] * 4 + ['b'] * 4
v_color = ['r'] * 5 + ['b'] * 5
small_mhg.draw(v_label=small_v_label, v_color=v_color, e_color=e_color)
```

```python
small_hg1_temp = dhg.Hypergraph(5, [(0, 1, 2), (2, 3), (0, 4)])
small_hg2_temp = dhg.Hypergraph(5, [(1, 3, 2), (0, 3), (2, 4)])
small_mhg_temp = MultilayerHypergraph(10, 2, layers_list=[small_hg1_temp, small_hg2_temp], prob_inner_layer_connect=0.7)
small_mhg_temp = small_mhg_temp.construct_multi_layer_hypergraph()
print(small_mhg_temp)
```

```python
small_v_label_temp = [str(i) for i in range(small_mhg_temp.num_v)]
e_color = ['r'] * 7 + ['b'] * 8
v_color = ['r'] * 5 + ['b'] * 5
small_mhg.draw(v_label=small_v_label_temp, v_color=v_color, e_color=e_color)
```
